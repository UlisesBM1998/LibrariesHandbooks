```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# MATPLOTLIB

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.figure(figsize = (12,4))
plt.plot(x,y, "red")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,x,"g--")
```

```python
fig = plt.figure(figsize=(12,4))
axes = fig.add_axes([0.0,0.0,0.8,0.8])

axes.plot(x,y, "black")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X vs Y")
```

```python
fig = plt.figure(figsize=(12,4))
axes1 = fig.add_axes([0.1,.1,.8,.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,"red")
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("X vs Y")

axes2.plot(y,x,"red")
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Y vs X")

plt.show()
```

```python
fig, axes = plt.subplots(nrows= 1, ncols = 2, figsize=(12,5), dpi=100)
axes[0].grid(True)
axes[1].grid(True)
for i in axes:
    i.plot(x,y)
    i.set_xlabel("X")
    i.set_ylabel("Y")
    
plt.show()
```

```python
fig.savefig("filename.png")
fig.savefig("filename.png", dpi=200)
```

```python
fig = plt.figure(figsize=(12,4))

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, x**2, "red", label = "X**2")
axes.plot(x,x**3, "green", label = "X**3")
axes.legend(loc=0, fontsize=30)

plt.show()
```

```python
fig = plt.figure(figsize=(12,4))

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x, x**2, color="#eeaa41", alpha=0.3 )
axes.plot(x,x**3, color="#52aa10" )

axes.plot(x,x**4, color="blue", ls = "-", lw = 3) # ls= ":" ls= ".-"
axes.plot(x,x**5, color="purple", marker = "o", markersize =10, markeredgecolor = "yellow", 
          markeredgewidth = 2, markerfacecolor= "brown") #market = "s", "+", "1" , "o"
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)

axes[1].plot(x,x**2,x,x**3)
axes[1].axis("tight")

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,50])
axes[2].set_xlim([2,5])
```

```python
plt.scatter(x,y, s=40, # tamaño de los puntos
               color='green', # color
               alpha=0.7, #transparencia de los puntos
               edgecolors='none')
```

```python
dataHist = np.random.randint(1,16,100)
plt.hist(dataHist, bins = 5,edgecolor="white", linewidth=0.7) #width rwidth
plt.show()
```

```python
dataBox = [np.random.normal(1,std,100) for std in range(1,4)]
plt.boxplot(dataBox, vert = True, patch_artist= True, flierprops={"color": "black", "markerfacecolor": "cyan", "markersize": 9},
                medianprops={"color": "white", "linewidth": 1.5},
                boxprops={"facecolor": "pink", "edgecolor": "white", "linewidth": 0.5},
                whiskerprops={"color": "magenta", "linewidth": 2.5},
                capprops={"color": "green", "linewidth": 3.5})
plt.show()
```

```python
heatArray=np.array([[1, 1, 2, 2, 3],  
            [2, 2, 3, 3, 5],  
            [1, 4, 2, 2, 3]]) 
heatCorr = np.corrcoef(heatArray)

fig, ax = plt.subplots()

im=ax.imshow(heatCorr, cmap = "magma")
fig.colorbar(im, ax=ax, label = "Correlations")
plt.show()
```

# SEABORN

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set_theme(style="darkgrid")
sns.set_style('white')
sns.set_style('ticks')
sns.set_context('poster',font_scale=4)
```

```python
sns.displot(tips["tip"], kde=True, bins=5)
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator = np.std)
plt.show()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.despine()
sns.despine(left=True)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow', orient='v', hue="smoker")
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```

```python
sns.jointplot(x= "total_bill", y="tip", data = tips, xlim=(0, 60), ylim=(0, 12),
                  color="m", height=7 , kind = "reg") #kind={ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
tipsPair = sns.PairGrid(tips)
tipsPair.map(plt.scatter)
```

```python
g = sns.PairGrid(tips)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
plt.figure(figsize=(20,12))
ax = sns.heatmap(flights, annot = True, linewidth = 0.5)
```

```python
sns.heatmap(tips.corr(), annot = True, linewidth = 0.5)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```
